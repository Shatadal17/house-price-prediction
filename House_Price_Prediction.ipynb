House Price Prediction

```python
# importing the libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import fetch_california_housing
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
# Loading the California Housing dataset into variable housing
housing = fetch_california_housing()
```

```python
# Creating a dataframe object and loading the California Housing data into df
df = pd.DataFrame(housing.data, columns=housing.feature_names)
df['Target'] = housing.target
```

```python
# Checking the datasets
df.head()
```

```python
# Checking the number of rows & columns in the dataset
df.shape
```

```python
# Checking for missing values
df.isnull().sum()
```

```python
# Statistical Measure of the dataset

df.describe()
```

```python
# Finding the correlation between various features of the dataset

correlation = df.corr()
```

```python
# Plotting a heat map for understanding correlation between the features

plt.figure(figsize=(10,10))
sns.heatmap(correlation, cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```

Splitting the data and Target

```python
x = df.drop(['Target'], axis = 1)
y = df['Target']
```

```python
print(x)
print(y)
```

Splitting data into Training and Testing Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Model Creation -- XGBoost Regression

```python
# Loading the model
model = XGBRegressor()
```

```python
# Training the model with x_train

model.fit(x_train, y_train)
```

Model Evaluation -- Prediction on training data

```python
# Finding the accuracy for prediction on training data

training_data_prediction = model.predict(x_train)
```

```python
print(training_data_prediction)
```

```python
# R squared error
score_1 = metrics.r2_score(y_train, training_data_prediction)

# Mean Absolute Error
score_2 = metrics.mean_absolute_error(y_train, training_data_prediction)

print("R squared error :", score_1)
print("Mean Absolute Error :", score_2)
```

Visualizations on Actual vs Predicted prices

```python
plt.figure(figsize=(14,8))
plt.scatter(y_train, training_data_prediction)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Prices vs Predicted Prices")
plt.show()
```

Predictions on Test data

```python
# Finding the accuracy for prediction on test data

test_data_prediction = model.predict(x_test)
```

```python
# R squared error
score_3 = metrics.r2_score(y_test, test_data_prediction)

# Mean Absolute Error
score_4 = metrics.mean_absolute_error(y_test, test_data_prediction)

print("R squared error :", score_3)
print("Mean Absolute Error :", score_4)
```
